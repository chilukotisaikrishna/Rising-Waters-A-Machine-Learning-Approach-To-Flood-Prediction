# Rising Waters — EDA
Exploratory analysis of flood occurrences and rainfall (India 1901–2015).

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path
%matplotlib inline
```

## Load data
Adjust paths if your dataset folder differs.

```python
data_dir = Path('..') / 'dataset'
flood_path = data_dir / 'flood dataset.xlsx'
rain_path = data_dir / 'rainfall in india 1901-2015.xlsx'

flood_df = pd.read_excel(flood_path)
rain_df = pd.read_excel(rain_path)

flood_df.head(), rain_df.head()
```

## Inspect columns and info

```python
flood_df.info()
rain_df.info()
```

## Quick EDA ideas
- Check missingness
- Distribution of rainfall over years/states
- Flood frequency by year/state
- Correlation between rainfall and flood occurrences

```python
fig, ax = plt.subplots(1, 2, figsize=(12, 4))
sns.histplot(rain_df.select_dtypes(include=[np.number]).iloc[:,0], ax=ax[0], kde=True)
ax[0].set_title('Rainfall numeric sample')
sns.countplot(data=flood_df, x=flood_df.columns[0], ax=ax[1])
ax[1].set_title('Flood target sample')
plt.tight_layout()
```
